## Multi-Linear Regression Model

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.random.rand(100, 2)  # 2 columns, meaning 2 different variables (X1 and X2)
coef = np.array([3, 5])  # Specifies the actual linear regression coefficients (3 for x1, 5 for x2)
y = 10 * np.random.rand(100) + np.dot(X, coef)  # y = 3X1 + 5X2. The dot function performs matrix multiplication, and this gives us the equation result.

lin_reg = LinearRegression()
lin_reg.fit(X, y)  # The model is trained using X1, X2, and y variables (similar to X_train and y_train)

# 3D plot generation
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")  # add_subplot(111): Adds a subplot to the figure / 111 means 1 row, 1 column, and the first plot
ax.scatter(X[:, 0], X[:, 1], y)  # In a normal regression model, only X and y would be shown. Here, the second X is shown as it represents the other variable.
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")

# Plotting the regression plane (Surface plot)
x1, x2 = np.meshgrid(np.linspace(0, 1, 10), np.linspace(0, 1, 10))
y_pred = lin_reg.predict(np.array([x1.flatten(), x2.flatten()]).T)  # T: transpose operation
ax.plot_surface(x1, x2, y_pred.reshape(x1.shape), alpha=0.6)  # Reshaped to match the shape of X1
plt.title("Multi-Linear Regression")
```

## Notes
np.meshgrid(): This function takes two 1D arrays and transforms them into a 2D coordinate grid (matrix form).

np.linspace(0, 1, 10): Generates 10 equally spaced numbers between 0 and 1.

x1.flatten() and x2.flatten(): These functions flatten the two 2D matrices, converting them into a single row.

np.array([...]).T: This combines (concatenates) the two vectors into a 2D matrix, which is used as the input data for the model's prediction.

NOTE: The reason for using flatten() --> The lin_reg.predict() function requires a 1D array (a row vector) for each data point, not a 2D input.

NOTE: The reason for using transpose (T)--> The predict() function of the model expects the input data (X) to be provided as one observation per row. By transposing, we swap rows and columns so that the X1 variable is treated as one row and the X2 variable is treated as a separate row for analysis.
