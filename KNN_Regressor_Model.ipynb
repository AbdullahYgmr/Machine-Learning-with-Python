## KNN-Regressor Model

## Prepare random datasets

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.sort(5 * np.random.rand(40,1), axis = 0)
y = np.sin(X).ravel()

## adding noise
y[::5] += 1 * (0.5 - np.random.rand(8))
plt.scatter(X,y)
```

## Select Regressor Model

```python
from sklearn.neighbors import KNeighborsRegressor
A = np.linspace(0,5,500)[: , np.newaxis]
```

```python
import matplotlib.pyplot as plt
for i, weight in enumerate(["uniform", "distance"]):
  knn = KNeighborsRegressor(n_neighbors= 5 , weights= weight)
  y_pred = knn.fit(X,y).predict(A)
  plt.subplot(2,1,i+1)
  plt.scatter(X,y, color = "lime" , label = "data")
  plt.plot(A,y_pred, color= "blue" , label = "predictions")
  plt.legend()
  plt.axis("tight")
  plt.title("KNN-Regressor Model = {}".format(weight))

plt.tight_layout()
plt.show()
```
